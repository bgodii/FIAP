```python
import pandas as pd
import numpy as np
import random
```

```python
def znorm(x: np.ndarray) -> np.ndarray:
    return (x - min(x))  / (max(x) - min(x))
```

```python
vetor_a = np.array(list(random.randint(1, 20) for i in range(20)))
vetor_b = np.array(list(random.randint(1, 20) for i in range(20)))
print(vetor_a, '\n', vetor_b)
```

```python
norm_vetor_a = znorm(vetor_a)
norm_vetor_b = znorm(vetor_b)
print(norm_vetor_a, '\n', norm_vetor_b)
```

```python
ab_matrix = np.array([vetor_a, vetor_b]).transpose()
norm_ab_matrix = np.array([norm_vetor_a, norm_vetor_b]).transpose()
print(ab_matrix, '\n', norm_ab_matrix)
```

```python
multiplied_array = np.array([ab_matrix * norm_ab_matrix])
multiplied_array
```

```python
ab_matrix.dot(norm_ab_matrix.transpose())
```
